## Modulo 1 - Importando datasets
Importando e Exportando Data em Python.
Começandoa Analisar Data em Python.
Python Packages para a Ciência de Dados

```
# Importando os pacote do Pandas
import pandas as pd
```

```
# Importando e lendo o Datasets
URL = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'
```

```
df = pd.read_csv(URL,header=None)
```

```
df.head()
```

```
# Manipulando o cabeçalho do DataFrame
headers = ['symboling','normalized-losses','make', 'fuel-type','aspiration', 'num-of-doors', 'bodystyle','drive-wheels','engine-location','wheel-base','length','width','height','curb-weight','engine-type','num-of-cylinders','engine-size','fuel-system','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price']
```

```
df.columns = headers
```

```
df.head()
```

```
# Exportando o Data Frame para o arquivo fisico
path = '/content/sample_data/automobile.csv'
```

```
df.to_csv(path)
```
